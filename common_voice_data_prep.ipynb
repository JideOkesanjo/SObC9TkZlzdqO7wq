## Loading The Datasets

```python
from datasets import load_dataset

# Load the dataset (not in streaming mode for this example)
dataset = load_dataset("mozilla-foundation/common_voice_13_0", "en", split="train")
```

## Collecting a Random Sample of the Data
Common voice was quite large, over 92 gigabytes of raw voice data. Due to constraints in processing power, the common voice data was subsampled to match the size of the smaller cloned voice dataset. This process aims to create a balanced dataset where the number of real voice samples closely matches the number of cloned voice samples.

### Random Selection of 1000 real voices from common voice data

```python
import numpy as np

# Assuming the dataset is already loaded and is not too large
total_samples = len(dataset)
indices = np.random.choice(total_samples, 1000, replace=False)  # Get 1000 unique indices
selected_samples = dataset.select(indices)
```

Copy and save the 1000 random voices into separate file to use for audio classification

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python

import os
import shutil
save_dir = '/content/drive/MyDrive/common_voice_13_0'
if not os.path.exists(save_dir):
  os.makedirs(save_dir)
for i , sample in enumerate(selected_samples):
  audio_path = sample['audio']['path']
  shutil.copy(audio_path, os.path.join(save_dir, f'sample_{i}.mp3'))
```
